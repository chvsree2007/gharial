```python
def convert_name_cell(val):
    if val == 'Newton':
        return 'Isaac Newton'
    return val

def convert_marks_percent_cell(val):
    if val == 'n.a' or val == 'not available':
        return None
    return val

df = pd.read_excel('test_data2.xlsx','Sheet1')
print(df)
print('\n')

df =pd.read_excel("test_data2.xlsx",'Sheet1',converters={
    'Name':convert_name_cell,
    'Marks':convert_marks_percent_cell,
    '%':convert_marks_percent_cell,
})
print(df)
#df.to_excel('new.xlsx','students',index=None)
df.to_excel('new.xlsx','students',index=None,startrow=2,startcol=1)
```
